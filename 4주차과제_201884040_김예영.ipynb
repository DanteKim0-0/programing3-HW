# 1. movie 데이터 크기 줄이기

```python
import pandas as pd
import numpy as np
```

```python
movie = pd.read_csv('programming3/data/movie.csv')
movie
```

```python
# movie.info()
```

```python
movie1 = movie.iloc[:, :11].fillna(0)
```

```python
movie1.info()
```

```python
movie1.memory_usage(deep=True)
```

```python
movie1.isna().sum()  # na 값이 있는지 확인
```

```python
movie1.describe().T
```

```python
np.iinfo('int32')
```

```python
np.iinfo('int16')
```

```python
movie1['num_critic_for_reviews'] = movie1['num_critic_for_reviews'].astype(np.int16)
movie1['duration'] = movie1['duration'].astype(np.int16)
movie1['director_facebook_likes'] = movie1['director_facebook_likes'].astype(np.int16)
movie1['actor_3_facebook_likes'] = movie1['actor_3_facebook_likes'].astype(np.int16)
movie1['actor_1_facebook_likes'] = movie1['actor_1_facebook_likes'].astype(np.int32)
movie1['gross'] = movie1['gross'].astype(np.int32)

movie1.memory_usage(deep=True)
```

 num_critic_for_reviews, duration 과 object를 제외한 나머지 메모리가 줄어든 것을 볼 수 있습니다.

```python
print('감독, 배우3 페북 좋아요 수 :', 9832 / 39328 *100, '%')
print('gross와 배우1 페북 좋아요 수 :', 19664 / 39328*100, '%') 
```

```python
movie1.select_dtypes(include=['object']).nunique()
```

```python
movie1['color'] = movie1['color'].astype('category')
movie1['genres'] = movie1['genres'].astype('category')
movie1['director_name'] = movie1['director_name'].astype('category')
movie1['actor_2_name'] = movie1['actor_2_name'].astype('category')
movie1['actor_1_name'] = movie1['actor_1_name'].astype('category')

movie1.memory_usage(deep=True)
```

```python
print('컬러/흑백영화 개수: ', 5190 / 306262 * 100, '%')
print('감독이름 개수: ', 245814 / 343388 * 100, '%')
print('장르 개수: ', 121102 / 379929 * 100, '%') 
print('배우2 이름 개수: ', 290652 / 346635 * 100, '%')
print('배우1 이름 개수: ', 224147 / 346599 * 100, '%')
```

# 2. college 데이터에서 도시별 최대 UGDS 값 구하기

INSTNM, CITY, UGDS

```python
colle = pd.read_csv('programming3/data/college.csv')
colle.head()
```

```python
colle2 = colle[['INSTNM', 'CITY', 'UGDS']]
colle2.head()
```

```python
# nlargest 사용
colle2.nlargest(80, 'UGDS').head()
```

```python
# sort_values 사용
(
colle[['INSTNM', 'CITY', 'UGDS']]
.sort_values(['UGDS'], ascending=[False])
)
```
