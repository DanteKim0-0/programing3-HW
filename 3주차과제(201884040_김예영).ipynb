# chunksize를 이용해 데이터 불러오기(최후 방법!)

```python
movies_iter = pd.read_csv('programing3/data/movie.csv',
                         dtype={
                            'num_critic_for_reviews' : np.float16,
                            'duration' : np.float16,
                            'director_facebook_likes' : np.float16,
                            'actor_3_facebook_likes' : np.float16,
                            'actor_1_facebook_likes' : np.float16,
                            'gross' : np.float16,
                            'num_voted_users' : np.int32,
                            'cast_total_facebook_likes' : np.int32,
                            'facenumber_in_poster' : np.float16,
                            'num_user_for_reviews' : np.float16,
                            'budget' : np.float16,
                            'title_year' : np.float16,
                            'actor_2_facebook_likes' : np.float16,
                            'imdb_score' : np.float16,
                            'aspect_ratio' : np.float16,
                            'movie_facebook_likes' : np.int32,
                            'color' : 'category',
                            'content_rating' : 'category'
                        },
                       chunksize=1000
                       )
def process(df):
    return f'processed {df.shape} shape of df {df.iloc[0, :]}'

for chunk in movies_iter:
    print(process(chunk))
```
